```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_json('https://ai-process-sandy.s3.eu-west-1.amazonaws.com/purge/deviation.json')
```

```python
df.head()
```

```python
Gt_corners = df['gt_corners']
Rb_corners = df['rb_corners']

mean_Gt_corners = Gt_corners.mean()
max_Gt_corners = Gt_corners.max()
min_Gt_corners = Gt_corners.min()

mean_Rb_corners = Rb_corners.mean()
max_Rb_corners = Rb_corners.max()
min_Rb_corners = Rb_corners.min()

print(f"Statistics for Gt_corners - Mean: {mean_Gt_corners}, Max: {max_Gt_corners}, Min: {min_Gt_corners}")
print(f"Statistics for Rb_corners - Mean: {mean_Rb_corners}, Max: {max_Rb_corners}, Min: {min_Rb_corners}")

plt.figure(figsize=(10, 6))
plt.plot(Gt_corners)
plt.title('Gt_corners')
plt.xlabel('Index')
plt.ylabel('Values')
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(Rb_corners)
plt.title('Rb_corners')
plt.xlabel('Index')
plt.ylabel('Values')
plt.show()
```

```python
Gt_corners = df['Gt_corners']
Rb_corners = df['Rb_corners']

fig_Gt = px.line(x=df.index, y=Gt_corners, title='Gt_corners')
fig_Gt.show()

fig_Rb = px.line(x=df.index, y=Rb_corners, title='Rb_corners')
fig_Rb.show()
```
